```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
scores=[]
scores.append(cross_val_score(SVC(kernel='linear',gamma='auto',C=1),iris.data,iris.target,cv=5))
scores.append(cross_val_score(SVC(kernel='rbf',gamma='auto',C=1),iris.data,iris.target,cv=5))
scores.append(cross_val_score(SVC(kernel='linear',gamma='auto',C=20),iris.data,iris.target,cv=5))
```

```python
scores
```

```python
import numpy as np
kernels=["linear","rbf"]
c=[1,10,20]
for kernels_itr in kernels:
  for c_itr in c:
    cross_vl_sc=cross_val_score(SVC(kernel=kernels_itr,gamma="auto",C=c_itr),iris.data,iris.target,cv=5)
    print("kernel: "+str(kernels_itr)+", C: "+str(c_itr)+" Score: "+str(np.average(cross_vl_sc)))
```

In Grid Search CV the computational cost is very high as it is calculating the score for each C as specified and on each datapoint. So when my dataset is very large then it would take a long time to get result.

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(SVC(gamma="auto"), {
    "kernel":["linear","rbf"],
    "C":[1,10,20]
},cv=5)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
Result=pd.DataFrame(clf.cv_results_)
Result
```

```python
Result[["param_C","param_kernel","mean_test_score"]]
```

```python
print("Best Score: "+str(clf.best_score_)+"  Best Parameter: "+str(clf.best_params_))
```

Randomized Search will take random combination of parameters

```python
from sklearn.model_selection import  RandomizedSearchCV
clf_random=RandomizedSearchCV(SVC(gamma="auto"), {
    "kernel":["linear","rbf"],
    "C":[1,10,20]
},cv=5, n_iter=2)
clf_random.fit(iris.data,iris.target)
Result_random=pd.DataFrame(clf_random.cv_results_)
Result_random[["param_C","param_kernel","mean_test_score"]]
```

Hyperparameter tunning for different models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params={
    "svm":{
        "model":SVC(gamma="auto"),
        "params":{
            "C":[1,10,20],
            "kernel":["linear","rbf"]
        }
    },
    "random_forest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10]
        }
    },

    "logisticregression":{
        "model": LogisticRegression(multi_class="auto",solver="liblinear"),
        "params":{
            "C":[1,5,10]
        }
    }
}
```

```python
scores=[]
for model_name,model_para in model_params.items():
  clf_models=GridSearchCV(model_para["model"],model_para["params"],cv=5,return_train_score=False)
  clf_models.fit(iris.data,iris.target)
  scores.append({
      "model":model_name,
      "Best_score":clf_models.best_score_,
      "Best_parameters":clf_models.best_params_
  })
Scores_models=pd.DataFrame(scores)
Scores_models
```
